```python
import pandas as pd

# Замените 'your_file.xlsx' на имя вашего файла
file_path = '1.xlsx'

# Чтение данных из первого листа
data = pd.read_excel(file_path)

# Предполагаем, что первые два столбца содержат X и Y
z = data.iloc[:, :2].values.tolist()

# Вывод результата
print(z)
```

```python
z_numeric = data.iloc[:, :2].apply(pd.to_numeric, errors='coerce').dropna().values
X, Y = z_numeric[:, 0], z_numeric[:, 1]

# Построение диаграммы рассеивания
plt.figure(figsize=(15, 10))
plt.title('Диаграмма рассеивания и регрессионные прямые')
# Вычисление коэффициентов линейной регрессии
slope_xy, intercept_xy = np.polyfit(X, Y, 1)
slope_yx, intercept_yx = np.polyfit(Y, X, 1)
# Построение диаграммы рассеивания и регрессионных прямых
# Регрессионная прямая для Y = f(X)
regression_line_xy = slope_xy * X + intercept_xy
plt.plot(X, regression_line_xy, color='red', label='Регрессионная прямая y = f(x)')
# Регрессионная прямая для X = f(Y)
regression_line_yx = slope_yx * Y + intercept_yx
plt.plot(regression_line_yx, Y, color='blue', label='Регрессионная прямая x = f(y)')

plt.scatter(X, Y, color='blue', alpha=0.7, edgecolor='k')
plt.title('Диаграмма рассеивания', fontsize=16)
plt.xlabel('X', fontsize=14)
plt.ylabel('Y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.gca().set_aspect('equal', adjustable='datalim')
plt.legend()
plt.show()
```
