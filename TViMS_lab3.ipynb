```python
X= [-4.50,-2.10,-1.58,-1.42,-1.04,-0.43,-0.35,0.22,0.42,0.89,1.08,1.88,1.94,2.09,2.14,2.47,2.49,2.51,2.57,2.71,2.73,3.10,3.25,3.27,3.86,4.00,4.22,4.27,4.37,4.84,5.11,5.28,5.51,5.59,5.94,6.23,6.41,6.42,6.76,7.20,7.27,7.34,7.44,8.30,9.02,9.04,9.86,10.65,12.00,17.23]
n = len(X)
X.sort()
print(X)
```

**Числовые характеристики:**

```python
F = sts.norm(7.9,np.sqrt(66))
print("математическое ожидание: ", F.mean()) # математическое ожидание.
print("дисперсия: ", F.var()) # дисперсия
print("среднее квадратичное отклонение: ", F.std()) # среднее квадратичное (стандартное) отклонение.
```

```python
F = sts.norm(4.17,4.017)
print("математическое ожидание: ", F.mean()) # математическое ожидание.
print("дисперсия: ", F.var()) # дисперсия
print("среднее квадратичное отклонение: ", F.std()) # среднее квадратичное (стандартное) отклонение.
```

**Доверительный интервал для математического ожидания:**

```python
stw._zconfint_generic (F.mean(), F.std()/np.sqrt(n), 0.05, "two-sided")
```

**Доверительный интервал для дисперсии:**

```python
# Квантили хи-квадрат распределения
chi2_left = sts.chi2.ppf(0.05 / 2, df=n - 1)
chi2_right = sts.chi2.ppf(1 - 0.05 / 2, df=n - 1)

# Границы доверительного интервала
confidence_interval_lower = (n - 1) * F.var() / chi2_right
confidence_interval_upper = (n - 1) * F.var() / chi2_left

print(f"({confidence_interval_lower}, {confidence_interval_upper})")
```
